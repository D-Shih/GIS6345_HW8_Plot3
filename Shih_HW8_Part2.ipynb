```python
#HW8 Part 2
# Create a Horizontal Bar chart for a subset of the attributes in a shapefile
# https://docs.digital.mass.gov/dataset/massgis-data-acute-care-hospitals
# Modified code from:
# https://matplotlib.org/gallery/lines_bars_and_markers/barh.html#sphx-glr-gallery-lines-bars-and-markers-barh-py
# https://towardsdatascience.com/how-to-filter-rows-of-a-pandas-dataframe-by-column-value-51996ea621f8
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hospgdf_1 = gpd.read_file("HOSPITALS_PT.shp") #type(hospitals) is geopandas.geodataframe.GeoDataFrame # len is 75
```

```python
# I want to only include the hospitals that have an Emergency Room and that are in the Central Mass Region
filtered_1 = hospgdf_1.ER_STATUS.str.contains("Y") #type(filteredHospitals) is pandas.core.series.Series #len is 75
hospgdf_2 = hospgdf_1[filtered_1] #len is 70
filtered_2 = hospgdf_2.CHIAREGION.str.contains("Central Massachusetts") #len is 70
```

```python
hospitalsWithER = hospgdf_2[filtered_2]  # len is 8
```

```python
hosp = hospitalsWithER["NAME"]  #type(hosp) is pandas.core.series.Series #len is 8
y_pos = np.arange(len(hosp))    #type(y_pos) is numpy.ndarray  #len is 8
bedCounts = hospitalsWithER["BEDCOUNT"]  #type(hosp) is pandas.core.series.Series #len is 8
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

ax.barh(y_pos, bedCounts)
ax.set_yticks(y_pos)
ax.set_yticklabels(hosp)
ax.set_xlabel('Bed Count')
ax.set_title('Bed Count for Hospitals with an ER in Central Mass')
```
